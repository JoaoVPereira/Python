# Estruturas de repetição - while

```
numero = 1
while numero < 6:
  print(numero)
  numero += 1
print('---')
```

```
numero = 5
while numero > 0:
  print(numero)
  numero -= 1
```

```
soma = 0
numero = 1
while numero < 6:
  soma += numero
  numero += 1
print(soma)
```

```
numero = 12
while numero < 1 or numero > 10:
  numero = int(input('Digite um número de 1 a 10: '))
```

```
#EXERCICIO 1
soma = 0;
for i in range (0, 5):
  nota = int(input("Digite a nota: "))
  soma = soma + nota;
media = soma/5;
print(media)
```

```
i=0
soma = 0
while i < 5:
  nota = int(input("Digite a nota: "))
  soma = nota+soma
  i += 1
media = soma/5
print(media)
```

```
#EXERCICIO 2 - for
numero_tabuada = int(input("Digite o numero da tabuada"))
for numero in range (0, 11):
  print(numero_tabuada, "x", numero, "=", numero_tabuada*numero)
```
