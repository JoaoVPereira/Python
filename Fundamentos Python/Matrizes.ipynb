```
#EXERCÍCIO 1
lista = []
for i in range(1, 6):
  lista.append(int(input("Digite o numero: ")))

numero = 0;
soma = 0;
while numero <=4:
  soma += lista[numero]
  numero+=1

print("Soma:", soma)
```

```
#EXERCÍCIO 3
matriz = np.array([[2, 3, 1], 
                   [4, 5, 7]])

soma = 0;

for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    print(matriz[i][j])  ##Percorre cada elemento
    soma += matriz[i][j]
print(soma)
```
