```
aluno = {'Pedro' : 8.0, 'Maria': 9.0, 'João': 8.0, 'Renato': 5.0}
texto = ''

for nome, nota in aluno.items():
  r = f'{nome},{nota}'
  texto +=  r
  print(f'{nome},{nota}')
  with open('Exercicio.txt', 'w') as ex:
      ex.writelines(texto)
```
