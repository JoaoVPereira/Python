# Orientação a Objetos

```
class Triangulo:
  def __init__(self, lado1, lado2, lado3, base, altura): #CONSTRUTOR - self É UM PADRÃO DO PYTHON
    self.lado1 = lado1
    self.lado2 = lado2
    self.lado3 = lado3
    self.base = base
    self.altura = altura
  
  def area(self):
    return (self.base * self.altura) / 2

  def tipo(self):
    if self.lado1 > self.lado2 + self.lado3:
      return 'não é um triângulo'
    elif self.lado1 == self.lado2 or self.lado1 == self.lado3 or self.lado1 == self.lado2:
      return 'triângulo isósceles'
    else:
      return 'outro tipo de triângulo'
```

```
t1 = Triangulo(5, 1, 3, 4, 3)
```

```
print(t1.lado1, t1.lado2, t1.lado3, t1.base, t1.altura)
```

```
t1.area()
```

```
t1.tipo()
```

```
t2 = Triangulo(8, 8, 8, 16, 9)
print(t2.lado1, t2.lado2, t2.lado3, t2.base, t2.altura)
```

```
t2.area()
```

```
t2.tipo()
```

```
#EXERCÍCIO 1

class Aluno():
  def __init__(self, nota1, nota2):
    self.nota1 = nota1;
    self.nota2 = nota2;

  def calculaMedia(self):
    media = (self.nota1+self.nota2)/2;
    print('Média é:', media)

  def mostraDados(self):
    print('A nota 1 é:', self.nota1);
    print('A nota 2 é:', self.nota2);

  def resultado(self):
    media = (self.nota1+self.nota2)/2
    if media >=6:
      print('Aprovado');
    else:
      print('Reprovado');
```

```
aluno1 = Aluno(5, 4)
aluno2 = Aluno(8, 6)
```

```
aluno1.calculaMedia()
aluno1.mostraDados()
aluno1.resultado()
```
